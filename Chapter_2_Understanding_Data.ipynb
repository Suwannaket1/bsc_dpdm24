````python
# prompt: open sheet number 3 of the excel file and select only the table part

import pandas as pd

# Assuming your Excel file is located at '/content/drive/MyDrive/bsc_dpdm24_data/export62_64.xlsx'
excel_file_path = '/content/drive/MyDrive/DPDM24/export62_64.xlsx'

try:
    # Read sheet number 3 (index 2)
    df_sheet3 = pd.read_excel(excel_file_path, sheet_name=2, header=None)  # Assuming no header row in the table

    # Identify the table (you might need to adjust the row and column indices based on your data)
    # Example: Assuming the table starts from row 5 and column 1, and ends at row 20 and column 8
    start_row = 5
    end_row = 20
    start_col = 1
    end_col = 8

    # Extract the table part
    table_data = df_sheet3.iloc[start_row:end_row, start_col:end_col]

    # Now, 'table_data' contains only the DataFrame of the table
    print(table_data)

except FileNotFoundError:
    print(f"Error: File not found at {excel_file_path}")
except ValueError:
    print(f"Error: Sheet with index 2 (or sheet name) not found in the Excel file.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
````
