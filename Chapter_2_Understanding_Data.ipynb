````python
# prompt: calculate distance matrix of the df

import pandas as pd
from scipy.spatial.distance import pdist, squareform

# Assuming 'df' is your DataFrame with columns 'X1' and 'X2'
# Example:
# data = [[1, 2], [-1, 4], [0, 3], [-3, -2]]
# df = pd.DataFrame(data, columns=['X1', 'X2'])

# Calculate the pairwise distances between all rows
distance_matrix = squareform(pdist(df[['X1', 'X2']]))

# Convert the distance matrix to a DataFrame
distance_df = pd.DataFrame(distance_matrix, index=df.index, columns=df.index)

distance_df
````

````python
import numpy as np

def calculate_distance_matrix(df):
  """
  Calculates the distance matrix of a DataFrame using the L1-norm.

  Args:
    df: A Pandas DataFrame where each row represents a data point.

  Returns:
    A NumPy array representing the distance matrix.
  """
  num_points = len(df)
  distance_matrix = np.zeros((num_points, num_points))

  for i in range(num_points):
    for j in range(i + 1, num_points):
      distance = np.sum(np.abs(df.iloc[i] - df.iloc[j]))  # L1-norm
      distance_matrix[i, j] = distance
      distance_matrix[j, i] = distance  # Distance matrix is symmetric

  return distance_matrix

# Example usage (assuming 'df' is your DataFrame):
distance_matrix = calculate_distance_matrix(df)
distance_matrix
````
